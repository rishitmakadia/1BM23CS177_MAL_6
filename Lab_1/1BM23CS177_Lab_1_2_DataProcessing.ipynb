```python
import pandas as pd
import numpy as np
```

```python
adult = pd.read_csv("sample_data/adult.csv")
adult.info()
```

```python
# Replace ? with NaN
adult.replace("?", np.nan, inplace=True)

# Fill missing values
for col in adult.columns:
    if adult[col].dtype == 'object':
        adult[col].fillna(adult[col].mode()[0], inplace=True)
    else:
        adult[col].fillna(adult[col].mean(), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in adult.select_dtypes(include='object').columns:
    adult[col] = le.fit_transform(adult[col])
```

```python
num_cols = adult.select_dtypes(include=['int64', 'float64']).columns

Q1 = adult[num_cols].quantile(0.25)
Q3 = adult[num_cols].quantile(0.75)
IQR = Q3 - Q1

adult = adult[~(
    (adult[num_cols] < (Q1 - 1.5 * IQR)) |
    (adult[num_cols] > (Q3 + 1.5 * IQR))
).any(axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
adult_minmax = adult.copy()
adult_minmax[num_cols] = minmax_scaler.fit_transform(adult[num_cols])
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
adult_standard = adult.copy()
adult_standard[num_cols] = std_scaler.fit_transform(adult[num_cols])
```
