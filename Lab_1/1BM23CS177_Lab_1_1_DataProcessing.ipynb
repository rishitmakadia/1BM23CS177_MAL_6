```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
diabetes = pd.read_csv("sample_data/diabetes.csv")
```

```python
diabetes.info()
```

```python
# Data Cleaning

from sklearn.preprocessing import LabelEncoder

# 1. Handling Missing Values (NUMERIC ONLY)
diabetes.fillna(diabetes.select_dtypes(include='number').mean(), inplace=True)
```

```python
# 2. Handling Outliers (IQR Method)

Q1 = diabetes.select_dtypes(include='number').quantile(0.25)
Q3 = diabetes.select_dtypes(include='number').quantile(0.75)
IQR = Q3 - Q1

diabetes = diabetes[~(
    (diabetes.select_dtypes(include='number') < (Q1 - 1.5 * IQR)) |
    (diabetes.select_dtypes(include='number') > (Q3 + 1.5 * IQR))
).any(axis=1)]
```

```python
# 3. Handling Categorical Data (Label Encoding)
le = LabelEncoder()
diabetes['CLASS'] = le.fit_transform(diabetes['CLASS'])
```

```python
# Data Cleaning

# Select numeric columns only
num_cols = diabetes.select_dtypes(include=['int64', 'float64']).columns

# Remove ID columns (optional but recommended)
num_cols = num_cols.drop(['ID', 'No_Pation'], errors='ignore')

# 1. Min-Max Scaling
minmax_scaler = MinMaxScaler()
diabetes_minmax = diabetes.copy()
diabetes_minmax[num_cols] = minmax_scaler.fit_transform(diabetes[num_cols])

# 2. Standard Scaling
std_scaler = StandardScaler()
diabetes_standard = diabetes.copy()
diabetes_standard[num_cols] = std_scaler.fit_transform(diabetes[num_cols])
```
