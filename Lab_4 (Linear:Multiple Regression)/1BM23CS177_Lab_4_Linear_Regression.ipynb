```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/canada_per_capita_income.csv')
```

```python
df.info();
```

```python

X = df[['year']]
y = df['per capita income (US$)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Get coefficient and intercept
m = model.coef_[0]
b = model.intercept_

# Print equation
print("Linear Regression Equation:")
print(f"y = {m:.4f} * x + {b:.4f}")

# Predict for 2020
predicted_income = model.predict([[2020]])
print("Predicted per capita income in 2020:", predicted_income[0])

# Plot
plt.scatter(X, y, color='red')
plt.plot(X, model.predict(X), color='blue')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Year vs Per Capita Income")
plt.show()
```

```python

X = df['year'].values
Y = df['per capita income (US$)'].values
N = len(X)

x_mean = np.mean(X)
y_mean = np.mean(Y)

xy_mean = np.mean(X * Y)
x2_mean = np.mean(X ** 2)

# Calculate slope (m)
m = (xy_mean - x_mean * y_mean) / (x2_mean - x_mean ** 2)

# Calculate intercept (b)
b = y_mean - m * x_mean

# Print equation
print("Linear Regression Equation:")
print(f"y = {m:.4f} * x + {b:.4f}")

# Predict for 2020
predicted_income = m * 2020 + b
print("Predicted per capita income in 2020:", predicted_income)
```

```python
df2 = pd.read_csv('/content/sample_data/salary.csv')
```

```python
df2.info()
```

```python
# Check missing values
print(df2.isnull().sum())

# Fill missing values with mean
df2['YearsExperience'].fillna(df2['YearsExperience'].mean(), inplace=True)

# Define X and y
X = df2[['YearsExperience']]
y = df2['Salary']

# Create model
model = LinearRegression()
model.fit(X, y)

# Get slope and intercept
m = model.coef_[0]
b = model.intercept_

# Print regression equation
print("Linear Regression Equation:")
print(f"Salary = {m:.4f} * YearsExperience + {b:.4f}")

# Predict salary for 12 years experience
predicted_salary = model.predict([[12]])
print("Predicted Salary for 12 years experience:", predicted_salary[0])
```
